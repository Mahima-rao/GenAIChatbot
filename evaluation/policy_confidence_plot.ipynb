```python
import json
import matplotlib.pyplot as plt

with open("policy_eval_results.json") as f:
    results = json.load(f)

confidences = [r["similarity"] for r in results]
labels = [f"Test {i+1}" for i in range(len(results))]
colors = ['green' if r["response_ok"] else 'red' for r in results]
```

```python
# 2. Plot
```

```python
plt.figure(figsize=(10, 5))
plt.bar(labels, confidences, color=colors)
plt.axhline(0.75, linestyle="--", color="gray", label="Confidence Threshold (0.75)")
plt.title("Policy Retrieval Confidence Scores")
plt.ylabel("Similarity")
plt.xlabel("Test Case")
plt.ylim(0, 1.05)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 1. Load the semantic results
import json
import matplotlib.pyplot as plt

with open("evaluation/policy_eval_semantic.json") as f:
    results = json.load(f)

scores = [r["semantic_score"] for r in results]
passed = [r["response_ok"] for r in results]

# 2. Plot histogram of scores
plt.figure(figsize=(8, 5))
plt.hist(scores, bins=10, edgecolor='black', color='skyblue')
plt.axvline(0.85, color='red', linestyle='--', label='Threshold (0.85)')
plt.title("Semantic Similarity Score Distribution")
plt.xlabel("Cosine Similarity")
plt.ylabel("Number of Responses")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np

plt.figure(figsize=(8, 4))
colors = ['green' if p else 'red' for p in passed]
x = list(range(len(scores)))

plt.bar(x, scores, color=colors)
plt.axhline(0.85, linestyle='--', color='gray', label='Threshold (0.85)')
plt.xticks(x, [f"Test {i+1}" for i in x], rotation=45)
plt.ylabel("Semantic Similarity")
plt.title("Per-Response Semantic Scores")
plt.legend()
plt.tight_layout()
plt.show()
```
